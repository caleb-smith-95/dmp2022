# Anti-joins

```python
def anti_join(keep, remove):
    outer_join = keep.merge(remove, how = 'outer', indicator = True)
    df = outer_join[~(outer_join._merge == 'both')].drop('_merge', axis = 1)
    df.reset_index(inplace = True) #Column specific to this project
    df2 = df.drop(columns = 'index') #Column specific to this project
    return(df2)
```

```python
rbs = anti_join(rush500, rushing_qbs)
rbs
```

```python
wrs = anti_join(rec500, receiving_rbs) #Don't need to "remove" receiving QB's because Terrelle Pryor is technically
                                       #considered a "receiving rb" using this criteria
wrs #This includes tight ends. Not sure how to avoid this, or how big of a deal this is
```

# Group by age and "normalize"

```python
pass_ages = pass1000.groupby('Age').agg({'Age': 'count'})
pass_ages.rename(columns = {'Age': 'Count'}, inplace = True)
pass_ages
```

```python
sum(pass_ages['Count'])
```

```python
len(pass1000)
```

```python
pass_ages['Freq'] = pass_ages['Count'] / len(pass1000)
pass_ages['Pos'] = 'QB'
pass_ages.reset_index(inplace = True)
pass_ages
```

```python
run_ages = rbs.groupby('Age').agg({'Age': 'count'})
run_ages.rename(columns = {'Age': 'Count'}, inplace = True)
run_ages['Freq'] = run_ages['Count'] / len(rbs)
run_ages['Pos'] = 'RB'
run_ages.reset_index(inplace = True)
run_ages
```

```python
sum(run_ages['Freq'])
```

```python
catch_ages = wrs.groupby('Age').agg({'Age': 'count'})
catch_ages.rename(columns = {'Age': 'Count'}, inplace = True)
catch_ages['Freq'] = catch_ages['Count'] / len(wrs)
catch_ages['Pos'] = 'WR'
catch_ages.reset_index(inplace = True)
catch_ages
```

```python
sum(catch_ages['Freq'])
```

```python
kick_ages = kick10.groupby('Age').agg({'Age': 'count'})
kick_ages.rename(columns = {'Age': 'Count'}, inplace = True)
kick_ages['Freq'] = kick_ages['Count'] / len(kick10)
kick_ages['Pos'] = 'K'
kick_ages.reset_index(inplace = True)
kick_ages
```

```python
sum(kick_ages['Freq'])
```

```python
combined_ages = pd.concat([pass_ages, run_ages, catch_ages, kick_ages], axis = 0)
combined_ages
```

```python
combined_ages['Percent'] = round(combined_ages['Freq']*100, 2)
combined_ages
```

# Hypothesis Testing

```python
ss.ttest_ind(rush500.Age, rec500.Age)
```

```python
ss.ttest_ind(pass1000.Age, kick10.Age)
```

```python
ss.f_oneway(pass1000.Age, rush500.Age, rec500.Age, kick10.Age)
```

The ANOVA test provides great evidence that position matters to some extent. In addition, age difference in the two most similar positions (QBs and Kickers) is significant at the 5% level of significance.
