# Import 1978 database from web, debug

```python
rush78 = pd.read_html("https://www.pro-football-reference.com/years/1978/rushing.htm#rushing")
comb_rush = rush78[0]
comb_rush.columns = comb_rush.columns.droplevel()
comb_rush['Year'] = 1978
comb_rush.head()
```

# Pull in rushing data from 1979-2021

```python
years = range(1979, 2022)
years[0]
```

```python
for year in years:
    url = f"https://www.pro-football-reference.com/years/{year}/rushing.htm#rushing"
    data = pd.read_html(url)
    season_rush = data[0]
    season_rush.columns = season_rush.columns.droplevel()
    season_rush['Year'] = year
    comb_rush = pd.concat([comb_rush, season_rush], ignore_index = True)
    time.sleep(1)
```
