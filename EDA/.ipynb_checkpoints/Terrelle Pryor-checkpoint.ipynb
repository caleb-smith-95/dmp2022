```python
import pandas as pd
import plotly.express as px
import numpy as np
%matplotlib inline
from itertools import repeat
from urllib.request import urlopen
import html5lib
import ssl
import time
```

```python
data_pass = pd.read_html(f"https://www.pro-football-reference.com/years/1932/passing.htm")
comb_pass = data_pass[0]

data_rec = pd.read_html(f"https://www.pro-football-reference.com/years/1932/receiving.htm")
comb_rec = data_rec[0]
```

```python
years = range(1933, 2022)

for year in years:
    
    url_pass = f"https://www.pro-football-reference.com/years/{year}/passing.htm"
    url_rec = f"https://www.pro-football-reference.com/years/{year}/receiving.htm"
    
    data_pass = pd.read_html(url_pass)
    data_rec = pd.read_html(url_rec)
    
    season_pass = data_pass[0]
    season_rec = data_rec[0]
    
    #season_kick.columns = season_kick.columns.droplevel()
    
    #season_kick['Year'] = year
    #season_rec['Year'] = year
    
    comb_pass = pd.concat([comb_pass, season_pass], ignore_index = True)
    comb_rec = pd.concat([comb_rec, season_rec], ignore_index = True)
    
    time.sleep(1)
```

```python
comb_pass
```

```python
comb_pass['Player'] = comb_pass['Player'].str.replace(r'[*+]', '', regex = True)
comb_pass
```

```python
comb_rec['Player'] = comb_rec['Player'].str.replace(r'[*+]', '', regex = True)
comb_rec
```

```python
real_pass = comb_pass.Player != 'Player'
new_pass = comb_pass[real_pass]
new_pass
```

```python
real_rec = comb_rec.Player != 'Player'
new_rec = comb_rec[real_rec]
new_rec
```

```python
passyds_new = pd.to_numeric(new_pass.Yds, errors = 'coerce').fillna(0)
passyds_new
```

```python
pass_names = new_pass.Player
pass_names
```

```python
recyds_new = pd.to_numeric(new_rec.Yds, errors = 'coerce').fillna(0)
recyds_new
```

```python
rec_names = new_rec.Player
rec_names
```

```python
pass_data = {'Name': pass_names, 'Yds': passyds_new}
pass_final = pd.DataFrame(pass_data)
pass1000 = pass_final[pass_final.Yds >= 1000]
pass1000
```

```python
rec_data = {'Name': rec_names, 'Yds': recyds_new}
rec_final = pd.DataFrame(rec_data)
rec1000 = rec_final[rec_final.Yds >= 1000]
rec1000
```

```python
pass1000_names = pd.DataFrame(pass1000['Name'].unique())
rec1000_names = pd.DataFrame(rec1000['Name'].unique())
rec1000_names.merge(pass1000_names, how = 'inner')
```
