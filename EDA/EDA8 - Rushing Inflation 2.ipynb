# Find people common to both lists

```python
pass_names = pd.DataFrame(pass1000['Name'].unique())
len(pass_names)
```

```python
rush_names = pd.DataFrame(rush500['Player'].unique())
len(rush_names)
```

```python
type(pass_names)
```

```python
rushing_qbs = pass_names.merge(rush_names, how = 'inner')
rushing_qbs = rushing_qbs.rename(columns = {0: 'Player'})
rushing_qbs
```

```python
rush_qb_list = pd.Series.to_list(rushing_qbs['Player'])
rush_qb_list
```

```python
rush500_v2 = rush500.copy().reset_index().drop(['index'], axis = 1)
rush500_v2
```

```python
rb = []
qb = []

for l in range(len(rush500_v2)):
    if (rush500_v2.Player[l] in rush_qb_list):
        rb.append(False)
        qb.append(True)
    else:
        rb.append(True)
        qb.append(False)
```

```python
rush500_rb = rush500_v2[rb].reset_index().drop('index', axis = 1)
rush500_rb
```

```python
rush500_qb = rush500_v2[qb].reset_index().drop('index', axis = 1)
rush500_qb.head(10)
```

```python
rb_fig = px.histogram(rush500_rb, x = 'Age')
rb_fig.update_yaxes(range = [0,300])
rb_fig.update_xaxes(range = [20, 45])
rb_fig.show()
```

```python
qb_fig = px.histogram(pass1000, x = 'Age')
qb_fig.update_yaxes(range = [0,300])
qb_fig.update_xaxes(range = [20, 45])
qb_fig.show()
```
