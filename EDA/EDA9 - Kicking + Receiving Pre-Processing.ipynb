# Pull in kicking and receiving data from 1979-2021

```python
years = range(1979, 2022)
years[0]
```

```python
for year in years:
    #url_kick = f"https://www.pro-football-reference.com/years/{year}/kicking.htm"
    url_rec = f"https://www.pro-football-reference.com/years/{year}/receiving.htm"
    
    #data_kick = pd.read_html(url_kick)
    data_rec = pd.read_html(url_rec)
    
    #season_kick = data_kick[0]
    season_rec = data_rec[0]
    
    #season_kick.columns = season_kick.columns.droplevel()
    
    #season_kick['Year'] = year
    season_rec['Year'] = year
    
    #comb_kick = pd.concat([comb_kick, season_kick], ignore_index = True)
    comb_rec = pd.concat([comb_rec, season_rec], ignore_index = True)
    
    time.sleep(1)
```

# Clean the data

```python
comb_rec.drop(columns = ['Y/Tgt', 'Tgt', 'Ctch%', '1D'], inplace = True)
comb_rec
```

```python
real_rec = comb_rec.Player != 'Player'
comb_rec = comb_rec[real_rec]
comb_rec
```

```python
comb_rec['Player'] = comb_rec['Player'].str.replace(r'[*+]', '', regex = True)
comb_rec
```
